# Process ETL

### Import libraries

```python
#import libraries
import pandas
import requests
from sqlalchemy import create_engine
#from access_server import user_n_password1, user_n_password2 
```

### Make request

```python
url = "https://api.jikan.moe/v3/search/anime?q=berserk"
response = requests.get(url)
```

### Status code

```python
response
```

### Convert response to json format

```python
#convert response to json
data_json = response.json()
```

### Get top 10 results on the list

```python
#get first 10 results on the list
data_anime = data_json["results"][:10]
data_anime
```

### Convert list to dataframe  

```python
dataframe = pandas.DataFrame(data_anime)
dataframe
```

### Drop columns

```python
#delete columns 
dataframe.drop(columns=["mal_id", "url", "image_url", "airing", "members"], inplace=bool(1))
```

### Create a new column: id_anime

```python
#create index column
dataframe["id_anime"] = dataframe.index
```

### Rename columns

```python
dataframe.rename(columns={"start_date":"release_date", "end_date":"finish_date"}, inplace=1)
```

### Change format date

```python
dataframe["release_date"] = pandas.to_datetime(dataframe["release_date"], format="%Y-%m-%d")
dataframe["finish_date"] = pandas.to_datetime(dataframe["finish_date"], format="%Y-%m-%d")
```

```python
dataframe.dtypes
```

### View

```python
dataframe
```

### Filter animes with the word Berserk in the title

```python
#filter just animes what contains berserk in the title
titles_with_the_word_berserk = dataframe["title"].str.contains("Berserk")
dataframe_berserk = dataframe.loc[titles_with_the_word_berserk]
```
